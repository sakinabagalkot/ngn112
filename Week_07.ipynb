```python
#Slide 72: Reading the data
import pandas as pd
Iris=pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',header=None)
Iris.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
Iris.head()
```

```python
#Slide 73: Numerical Summaries
x=Iris['class'].value_counts()
print('This the value count per class\n', x)
print('\n\n This is the output of the data.info() command\n')
Iris.info()
```

```python
#Slide 76 - The describe function
print('\n\n This is the output of the data.describe() command\n')
print(Iris.describe())
```

```python
#Slide 75 - The describe function with include all
print("\n\n This is the output of the data.describe(include='all') command\n")
print(Iris.describe(include='all'))
```

```python
Iris['sepal_length'].min()
```

```python
#Slide 74: Manually generating numerical summary
from pandas.api.types import is_numeric_dtype
for col in Iris.columns:
  if is_numeric_dtype(Iris[col]):
    print('%s:' % (col))
    print('\t Mean = %.3f' % Iris[col].mean())
    print('\t Standard deviation = %.2f' % Iris[col].std())
    print('\t Minimum = %.2f' % Iris[col].min())
    print('\t Maximum = %.2f' % Iris[col].max())
```

```python
#slide 77: Visual Summaries Histogram using matplotlib
import matplotlib.pyplot as plt
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')
plt.ylim(0,40)
Iris['sepal_length'].hist(bins=8)
plt.show()
```

```python
#Slide 77: Visual summaries Histogram using Seaborn
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='darkgrid')
plt.figure(figsize=(5,4))
ax=sns.histplot(data=Iris['sepal_length'], bins=8)
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Frequency')
ax.set_title('Histogram of Sepal Length')
```
